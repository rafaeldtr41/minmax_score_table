The min max algorithm is used for rivals local searching on adversaries. It has a lot of use in videogame area, bc the
AI use this algorithm to build adversaries against the player. Basically  the algoritm walks trough a t tree looking for
best and worst value bc these are the best play for the ai and the worst play for the opponent. Usually they obstruct some 
results on them bc when you play you dont always want to lose. 

On this case we are looking for the best, worst outcome for two teams in order to analyze if on their best outcome who have 
more chance of winning.

In order to do this we going to evaluate every game of the team. we going to calc a weight for every team, so we have two weight for every 
game. Whats the rules in order to create the weight?

>Weight is a measure between 1 and 0

> We going to represent, using fuzzy logic how much a score obtained represent a team won.

 Fuzzy logic apply discrete mathematics in data that can't be represented like absolute true or absolute false. In this example, the positive diference
 between the gol score on the two teams represent who much close is the team for a absolute victory. the winning team is awarded and the losing
 team is punished, bc a score like 1-0 is a victory, but a tie like 1-1 has to be less weight than a victory bc here the advantage correspond to the away team
 bc away teams gols worth double in case of tie.

 A score like 5-1 have to be more weight than a 1-0 both are victories but, in the first case the diference are more big and it has more chance of absolute victory.

>If a team won, the team is prized

>If a team lose, the team is punished



```python
import math

score = 1
score1 = 0# asuming score1 is away team
# input.data_structure line 202

val = math.sqrt(max(float(score), float(score1)*1.2)*0.1) - math.pow(min(float(score),(float(score1)*1.2))*0.1, 2)
val1 = math.pow(min(float(score),(float(score1)*1.2))*0.1, 2)

print(val)
print(val1)
```
